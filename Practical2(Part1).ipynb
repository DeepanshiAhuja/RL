```python
import numpy as np
```

```python
x = np.random.randint(0,50,1000)
x
```

```python
# positive correlation with some noise
y=x+np.random.normal(0,10,1000)
y
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

```python
z= 100-np.random.randint(0,10,1000)
z
```

```python
np.corrcoef(x,z)
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')

plt.scatter(x,z)
plt.show()
```

```python
import pandas as  pd
df=pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b']=df['a']+np.random.normal(0,10,1000) #positively correlated with 'a'
df['c']=100-df['a']+np.random.normal(0,5,1000) # negatively correlated with 'a'
df['d']=np.random.randint(0,50,1000) #not correlated with 'a'

df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
